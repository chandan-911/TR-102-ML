```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
userage=np.array([20,30,34,40,50,70])
salleries=np.array([10,30,20,40,50,67])
```

```python
plt.plot(userage,salleries)
```

```python
plt.plot(userage,salleries,marker=".",ms=12)
```

```python
plt.plot(userage,salleries,marker=".",ms=12,mec="red",mfc="yellow")
```

```python
plt.plot(userage,salleries,marker=".",ms=12,mec="red",mfc="yellow",
        color="green",
        linewidth=2,
        linestyle="dashdot")
```

```python
plt.plot(userage,salleries,marker=".",ms=12,mec="red",mfc="yellow",
        color="green",
        linewidth=2,
        linestyle="dashdot")
plt.xlabel("emp age")
plt.ylabel("emp saleries")
plt.title("employee info",loc="left",color="red")#center and right also
```

```python
plt.plot(userage,salleries,marker=".",ms=12,mec="red",mfc="yellow",
        color="green",
        linewidth=2,
        linestyle="dashdot")
plt.xlabel("emp age")
plt.ylabel("emp saleries")
plt.title("employee info",loc="left",color="red")#center and right also
plt.grid(color="lightgreen")
```

```python
plt.plot(userage,salleries,marker=".",ms=12,mec="red",mfc="yellow",
        color="green",
        linewidth=2,
        linestyle="dashdot")
plt.xlabel("emp age")
plt.ylabel("emp saleries")
plt.title("employee info",loc="left",color="red")#center and right also
plt.grid(color="lightgreen")


# plt.annotate(xy=(40,40),text="hello")

for i in range(len(userage)):
    plt.annotate(xy=(userage[i],salleries[i]),text=f"{userage[i]},{salleries[i]}")
```

```python
month=np.array([1,2,3,4,2,4,5,6,7])
sales=np.array([200,300,233,500,600,455,322,445,667])
profit = [200,300,400,500,600,700,233,450,670,960,897]

coor = ["red", "green", "yellow", "pink", "blue", "orange", "gray", "lightgreen", "brown"]
plt.scatter(month, sales, sizes=profit,color=coor)
```

```python
plt.scatter(month, sales, sizes = profit,color=coor)
for i in range(len(month)):
    plt.annotate(xy=(month[i],sales[i]),text=f"{month[i]},{sales[i]}")
```
