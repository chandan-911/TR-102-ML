```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("placements_new.csv")
```

```python
data.head()
```

```python
x=data["cgpa"].values.reshape(200,1)
y=data["package"].values.reshape(200,1)
```

```python
# x=data.iloc[0:,:1]
# y=data.iloc[0:,1:2]
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,color="green")
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.title("Placements of students")
```

```python
# train_test_split
#     to break data in 2 parts
# 1. training data --> to train model
# 2. testing data --> to test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
# 20% data will used for test
```

```python
# x --> (200,1) 
# xtrain --> 80% (160)
# xtest --> 20%  (40)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
plt.scatter(xtrain,ytrain,color="green",label=("training data"))
plt.scatter(xtest,ytest,color="red",label=("testing data"))
plt.legend()
```

# self work

```python
from sklearn.linear_model import LinearRegression 
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
ypred
```

```python
ytest
```

```python
plt.scatter(xtrain,ytrain,color="lightblue",label="model Training")
plt.plot(xtest,ypred,color="green",label="model Prediction",linestyle="dashdot")
plt.grid()
plt.legend()
plt.xlabel("Independent variable")
plt.ylabel("Dependent variable")
plt.title("SLR Model")
```
